```python
from llama_cpp import Llama

llm = Llama.from_pretrained(
    repo_id="prism-ml/Bonsai-27B-gguf",
    filename="Bonsai-27B-Q1_0.gguf",   # 3.8 GB model
    verbose=False,
    n_ctx=4096,
)

headline = "India launches new AI research initiative"

prompt = f"""
You are a professional news journalist.

Write a well-structured news article (250-300 words) based on the following headline.

Headline:
{headline}

Article:
"""

output = llm(
    prompt,
    max_tokens=350,
    temperature=0.7,
    top_p=0.95,
    top_k=20,
    stop=["</s>"]
)

print(output["choices"][0]["text"])
```
